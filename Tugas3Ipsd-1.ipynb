**menghitung Mean**

```python
mean_values = data.mean(numeric_only=True)
print(mean_values)
```

```python
# mean ini menggunakan perhitungan dengan cara looping. 
mean_values = {}

for column in data.select_dtypes(include=[np.number]).columns:
    mean_values[column] = data[column].mean()

# Menampilkan hasil
for column, mean in mean_values.items():
    print(f"Rata-rata untuk kolom '{column}': {mean}")
 
```

kesimpulannya nilai mean antara langsung dan looping memiliki nilai yang sama.

**heatmap**

```python
# Mengambil hanya kolom numerik
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Menghitung matriks korelasi
correlation_matrix = numerical_data.corr()

# Membuat heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, cbar_kws={"shrink": .8})

# Menambahkan judul
plt.title('Heatmap Korelasi', fontsize=16)
plt.show()
```

**Jika dekan PT ingin mengetahui apakah ada relationship antara age dengan OUTPUT study satisfaction**

```python
# Menghitung korelasi antara usia dan kepuasan studi
correlation = data['age'].corr(data['study_satisfaction'])
print(f"Korelasi antara Usia dan Kepuasan Studi: {correlation:.2f}")

# Membuat scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='study_satisfaction', data=data)
plt.title('Hubungan antara Usia dan Kepuasan Studi', fontsize=16)
plt.xlabel('Usia', fontsize=14)
plt.ylabel('Kepuasan Studi', fontsize=14)
plt.grid(True)
plt.show()
```
